## Data Overview

We begin with a quick look at the dataset's shape, class distribution, and data quality before diving into the analysis.

## Challenge 1: Explore - Which Jobs Are Most Likely to Subscribe?

We calculate the subscription rate (proportion of customers who said "yes") for each job type to identify which professions respond best to the campaign.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

job_sub_rate = (bm.groupby('job')['y']
                .apply(lambda x: (x == 'yes').mean() * 100)
                .sort_values(ascending=False))

fig, ax = plt.subplots(figsize=(12, 5))
colors = ['#1565C0' if i < 3 else '#90CAF9' for i in range(len(job_sub_rate))]
job_sub_rate.plot(kind='bar', ax=ax, color=colors, edgecolor='white')
ax.set_title('Term Deposit Subscription Rate by Job Type', fontsize=14, fontweight='bold')
ax.set_xlabel('Job Type', fontsize=12)
ax.set_ylabel('Subscription Rate (%)', fontsize=12)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.tick_params(axis='x', rotation=45)
for i, v in enumerate(job_sub_rate):
    ax.text(i, v + 0.3, f'{v:.1f}%', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

print('Top 3 job types with the highest subscription rates:')
for job, rate in job_sub_rate.head(3).items():
    print(f'  {job}: {rate:.1f}%')
```

## Challenge 2: Visualize - Subscriptions by Month

We examine how the number of term deposit subscriptions varies across calendar months. This helps identify peak periods and can inform future campaign scheduling.

```python
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun',
               'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
month_subs = (bm[bm['y'] == 'yes']
              .groupby('month')
              .size()
              .reindex(month_order)
              .fillna(0))

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(range(len(month_subs)), month_subs.values, color='#FF7043', edgecolor='white')
ax.set_xticks(range(len(month_subs)))
ax.set_xticklabels([m.capitalize() for m in month_order], rotation=45)
ax.set_title('Number of Term Deposit Subscriptions by Month', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Number of Subscriptions', fontsize=12)
for bar, val in zip(bars, month_subs.values):
    if val > 0:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10,
                str(int(val)), ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

peak_month = month_subs.idxmax()
print(f'Peak subscription month: {peak_month.capitalize()} ({int(month_subs.max())} subscriptions)')
```

## Challenge 3: Analyze - Impact of Contact Frequency on Subscription

We explore whether contacting a customer more times during the campaign increases or decreases their likelihood of subscribing. More calls may reflect persistence, but could also indicate customer fatigue.

```python
camp = bm.copy()
camp['subscribed'] = (camp['y'] == 'yes').astype(int)
camp['campaign_bin'] = camp['campaign'].clip(upper=10)

campaign_rate = camp.groupby('campaign_bin').agg(
    subscription_rate=('subscribed', lambda x: x.mean() * 100),
    count=('subscribed', 'count')
).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.plot(campaign_rate['campaign_bin'], campaign_rate['subscription_rate'],
         'o-', color='#1565C0', linewidth=2.5, markersize=8, label='Subscription Rate')
ax2.bar(campaign_rate['campaign_bin'], campaign_rate['count'],
        alpha=0.3, color='#90CAF9', label='Number of Customers')

ax1.set_xlabel('Number of Contacts During Campaign (capped at 10+)', fontsize=12)
ax1.set_ylabel('Subscription Rate (%)', fontsize=12, color='#1565C0')
ax2.set_ylabel('Number of Customers', fontsize=12, color='gray')
ax1.set_title('Impact of Contact Frequency on Subscription Likelihood', fontsize=14, fontweight='bold')
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
plt.tight_layout()
plt.show()

rate_1 = campaign_rate.loc[campaign_rate['campaign_bin'] == 1, 'subscription_rate'].values[0]
rate_5plus = campaign_rate.loc[campaign_rate['campaign_bin'] >= 5, 'subscription_rate'].mean()
print(f'Subscription rate after 1 contact: {rate_1:.1f}%')
print(f'Average subscription rate after 5+ contacts: {rate_5plus:.1f}%')
```

## Data Preprocessing

Before modelling, we prepare the data:

- **Target encoding**: convert `y` (yes/no) to binary (1/0)
- **Leakage prevention**: drop `duration` since call duration is only known after the call ends, making it unavailable at prediction time
- **Feature engineering**: convert `pdays` (999 = never contacted) to a binary `previously_contacted` flag
- **Categorical encoding**: one-hot encode all categorical variables
- **Train/test split**: 80% training, 20% test, stratified to preserve class ratios
- **Scaling**: StandardScaler applied for Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df = bm.copy()

# Encode target variable
df['y'] = (df['y'] == 'yes').astype(int)

# Drop duration: only known after the call ends, so it would cause data leakage
df = df.drop(columns=['duration'])

# Engineer binary feature from pdays (999 means the client was never previously contacted)
df['previously_contacted'] = (df['pdays'] != 999).astype(int)
df = df.drop(columns=['pdays'])

# One-hot encode all categorical features
cat_cols = df.select_dtypes(include='object').columns.tolist()
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

X = df_encoded.drop(columns=['y'])
y_target = df_encoded['y']

X_train, X_test, y_train, y_test = train_test_split(
    X, y_target, test_size=0.2, random_state=42, stratify=y_target
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f'Features: {X_train.shape[1]}')
print(f'Training samples: {X_train.shape[0]:,}')
print(f'Test samples: {X_test.shape[0]:,}')
print(f'Subscription rate (train): {y_train.mean():.1%}')
imbalance = (y_train == 0).sum() / (y_train == 1).sum()
print(f'Class imbalance: {imbalance:.1f}:1 (no:yes)')
```

## Predictive Modeling

The dataset is imbalanced (roughly 11% subscription rate). We use `class_weight='balanced'` in both models to penalise misclassification of the minority class, and evaluate using ROC-AUC alongside precision and recall.

**Models compared:**
- **Logistic Regression**: a linear, interpretable baseline
- **Random Forest**: a non-linear ensemble that can capture complex feature interactions

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

lr = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)
y_prob_lr = lr.predict_proba(X_test_scaled)[:, 1]

print('=== Logistic Regression ===')
print(classification_report(y_test, y_pred_lr, target_names=['No Subscription', 'Subscription']))
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob_lr):.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]

print('=== Random Forest ===')
print(classification_report(y_test, y_pred_rf, target_names=['No Subscription', 'Subscription']))
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob_rf):.4f}')
```

### Model Comparison: ROC Curves

The ROC-AUC metric measures a model's ability to distinguish between subscribers and non-subscribers across all classification thresholds. A score of 1.0 is perfect; 0.5 equals random guessing.

```python
from sklearn.metrics import roc_curve, roc_auc_score

fig, ax = plt.subplots(figsize=(8, 6))

for name, y_prob in [('Logistic Regression', y_prob_lr), ('Random Forest', y_prob_rf)]:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob)
    ax.plot(fpr, tpr, linewidth=2, label=f'{name} (AUC = {auc:.3f})')

ax.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
ax.set_xlabel('False Positive Rate', fontsize=12)
ax.set_ylabel('True Positive Rate (Recall)', fontsize=12)
ax.set_title('ROC Curves: Model Comparison', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## What Drives Subscription? - Feature Importance

Feature importances from the Random Forest model show which variables most reduce prediction error, revealing the primary drivers of term deposit subscription.

```python
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=True).tail(15)

fig, ax = plt.subplots(figsize=(10, 8))
importance_df.set_index('Feature')['Importance'].plot(
    kind='barh', ax=ax, color='#1565C0', edgecolor='white')
ax.set_title('Top 15 Most Important Features (Random Forest)', fontsize=14, fontweight='bold')
ax.set_xlabel('Feature Importance', fontsize=12)
plt.tight_layout()
plt.show()

print('Top 10 most important features:')
top10 = importance_df.tail(10).iloc[::-1]
for _, row in top10.iterrows():
    feat = row['Feature']
    imp = row['Importance']
    print(f'  {feat}: {imp:.4f}')
```

## Conclusions

### Key Findings

**Exploratory Analysis:**
- **Students and retirees** have the highest subscription rates. These groups are most open to term deposit offers.
- **May** sees the highest raw volume of subscriptions (driven by campaign scale), while **March, September, October, and December** tend to have stronger conversion rates.
- Subscription rates **decline sharply after the 2nd or 3rd contact**. Over-contacting customers is counterproductive.

**Predictive Model:**
- Both Logistic Regression and Random Forest significantly outperform random chance.
- The Random Forest achieves a higher ROC-AUC, capturing non-linear interactions between features.
- Key predictors include the **euribor 3-month rate**, **number of employees** (nr.employed), **consumer confidence index**, and **previous campaign outcome**.

### Recommendations

1. **Target students and retirees**: these segments show the highest willingness to subscribe.
2. **Schedule campaigns in March, September-October, and December**: conversion rates are higher in these months.
3. **Limit contacts to 2 per customer per campaign**: additional calls yield diminishing returns.
4. **Prioritise customers with a successful prior campaign outcome**: a positive previous interaction is a strong predictor.
5. **Monitor macroeconomic conditions**: lower euribor rates and higher consumer confidence correlate with greater subscription likelihood; intensify campaigns during these windows.
